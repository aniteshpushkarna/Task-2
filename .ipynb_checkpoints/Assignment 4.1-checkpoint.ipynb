# Which team has won/lost maximum number of matches while chasing?

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
path = 'ipldata/matches.csv'
file_df = pd.read_csv(path,parse_dates=['date'])
```

```python
df = file_df[['toss_winner','toss_decision','winner','result']].dropna().query('result == "normal"')
```

```python
win = df.query('(toss_winner == winner and toss_decision == "field") or (toss_winner != winner and toss_decision == "bat")')
win_count = Counter(win['winner'])
```

```python
lost = df.query('(toss_winner != winner and toss_decision == "field") or (toss_winner == winner and toss_decision == "bat")')
lost_count = Counter(lost['winner'])
```

```python
ticks = range(len(win_count))
height = list(win_count.values())
tick_label = list(win_count.keys())
plt.barh(ticks,height,tick_label=tick_label,color=['r','g','b','m'])
```

```python
ticks = range(len(lost_count))
height = list(lost_count.values())
tick_label = list(lost_count.keys())
plt.barh(ticks,height,tick_label=tick_label,color=['r','g','b','m'])
```

```python
print(f'Most Won team while chasing is {win_count.most_common()[0]}')
print(f'Most Lost team while chasing is {lost_count.most_common()[0]}')
```

# Which toss decision is best for each ground?

```python
ground_df = file_df[['toss_winner','winner','toss_decision','venue']]
```

```python
def field_bat(x):
    f = list(x).count('field')
    b = list(x).count('bat')
    if f < b:
        return 'Bat'
    else:
        return 'Field'
```

```python
toss_decision = ground_df.query('toss_winner == winner')[['venue','toss_decision']]
toss_decision.groupby('venue').agg(field_bat)
```

# Which player has won maximum number of MoM awards while chasing?

```python
mom_df = file_df[['player_of_match','win_by_runs','win_by_wickets']]
```

```python
chasing = mom_df.query('win_by_wickets > 0')
chasing.groupby('player_of_match')['player_of_match'].agg('count').idxmax()
```

# Which batsman has played maximum number of dot balls

```python
p = 'ipldata\deliveries.csv'
deliveries = pd.read_csv(p)
```

```python
deliveries.query('total_runs == 0').groupby('batsman')['batsman'].agg('count').idxmax()
```

# Which batsmen came to crease but never got a chance to face a ball

```python
def no_ball(x):
    bat = set(x['batsman'])
    run = set(x['non_striker'])
    return run.intersection(bat)
```

```python
d1 = deliveries[['match_id','batsman','non_striker']]
d = d1.groupby('match_id').agg(no_ball).head()[['non_striker']]
for i in list(d['non_striker']):
    for j in i:
        print(j)
```
