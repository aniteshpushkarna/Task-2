```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'delhi-weather-data/testset.csv'
weather = pd.read_csv(path,parse_dates=['datetime_utc'])
```

# Data Cleaning

```python
weather.drop(['_precipm'],axis=1,inplace=True)
weather['_heatindexm'].fillna(weather['_heatindexm'].mean(),inplace=True)
weather['_vism'].fillna(weather['_vism'].mean(),inplace=True)
weather['_wspdm'].fillna(weather['_wspdm'].mean(),inplace=True)
weather.tail()
```

# Make a plot of different numeric features like rain, snow, dewptm, etc. over the duration of a month or year or day.

```python
data = pd.DataFrame()
data['Date'] = pd.to_datetime(weather['datetime_utc'])
data['month'] = data['Date'].dt.month
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4,figsize=(25,25))

month_mean = weather.groupby(data.month).mean()

month_mean._dewptm.plot(ax=axes[0,0],title='dewptm')
month_mean._fog.plot(ax=axes[0,1],title='fog')
month_mean._hail.plot(ax=axes[0,2],title='hail')
month_mean._heatindexm.plot(ax=axes[0,3],title='heat index')
month_mean._hum.plot(ax=axes[1,0],title='hum')
month_mean._pressurem.plot(ax=axes[1,1],title='pressurem')
month_mean._rain.plot(ax=axes[1,2],title='rain')
month_mean._snow.plot(ax=axes[1,3],title='snow')
month_mean._tempm.plot(ax=axes[2,0],title='tempm')
month_mean._thunder.plot(ax=axes[2,1],title='thunder')
month_mean._tornado.plot(ax=axes[2,2],title='tornado')
month_mean._wdird.plot(ax=axes[2,3],title='wdird')
month_mean._wgustm.plot(ax=axes[3,0],title='wgustm')
month_mean._vism.plot(ax=axes[3,1],title='vism')
month_mean._windchillm.plot(ax=axes[3,2],title='windchillm')
month_mean._wspdm.plot(ax=axes[3,3],title='wspdm')
```
